```python
#description: This is a data cleaning project done on SAT results from 460 schools in the city of New York.
#74 (16%) of the schools had no entries. 
#I used Python's Pandas library to normalize text, and fill missing entries with averages.
```

```python
import pandas as pd
import numpy as np
```

```python
#setting df to read
df = pd.read_csv("SAT__College_Board__2010_School_Level_Results.csv",encoding="latin")
```

```python
#count of total entries: 460
df.count()
```

```python
#count of non-entries: 74
#74/460=.16, only 84% available
df.isnull().sum()
```

```python
#reformatting header and school_name to remove spaces/case differences
df.columns=df.columns.str.replace(" ","_")
df.columns=df.columns.str.upper()
df['SCHOOL_NAME']=df['SCHOOL_NAME'].str.upper()
df.head()
```

```python
#fills NaN's with average of column 
df_mean=df.fillna(df.mean())
round(df_mean,0)
```
